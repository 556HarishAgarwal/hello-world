```python
# import python libraries after downloading them
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
df = pd.read_csv('/content/data.csv')
```

```python
sns.countplot(df['diagnosis'], label = 'count')
```

```python
sns.pairplot(df.iloc[:,1:7] , hue = 'diagnosis')
```

```python
plt.figure(figsize = (10,10))
sns.heatmap(df.iloc[:,1:11].corr(),cmap="YlGnBu",annot=True,fmt='.0%')
```

```python
X=df["diagnosis"]
Y=df["radius_mean"]
plt.scatter(X,Y)
plt.title("Scatter Plot")
plt.xlabel("Working Experience(year)")
plt.ylabel("Annual Wage")
plt.show()
```

```python
X_mean = df["radius_mean"]
Y_mean = df["texture_mean"]
average = X_mean[Y_mean == 0].mean(axis=0)
plt.bar(df.area_se,average)
plt.title("Bar chart Setosa Averages")
plt.ylabel("Average")
plt.show()
```

```python
bins = 20
X_mean = df["radius_mean"]
Y_mean = df["texture_mean"]
# average = X_mean[Y_mean == 0].mean(axis=0)
plt.hist(X_mean,bins)
plt.title("Histogram chart ")
plt.xlabel("names")
plt.ylabel("Average")
plt.show()
```

```python
# bins = 20
X_mean = df["radius_mean"]
Y_mean = df["texture_mean"]
# average = X_mean[Y_mean == 0].mean(axis=0)
plt.boxplot(X_mean,labels=[df.area_mean[0]])
plt.title("Boxplot chart ")
# plt.xlabel("names")
plt.ylabel("Data")
plt.show()
```

```python
sns.kdeplot(x = 'texture_mean', data = df, color = 'black')
```

```python
sns.displot(x='texture_mean', kde = True, bins = 5, data = df)
```

```python
sns.jointplot(x='texture_mean', y = 'area_mean', data = df, kind = 'reg')
```

```python
sns.jointplot(x = 'area_mean', y = 'texture_mean', data = df, hue = 'radius_mean')
```
